```python
import sklearn.datasets as datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load the Iris dataset
FlowerData = datasets.load_iris()

# Create a DataFrame with the dataset's features, ensuring column names are consistent
df = pd.DataFrame(FlowerData.data, columns=[name.lower() for name in FlowerData.feature_names])

# Get the target values
y = FlowerData.target

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=42)

# Train a Decision Tree classifier
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
# Ensure the feature names in x_test match those used during training
x_test.columns = [name.lower() for name in x_test.columns]

# Predict using the trained model
predictions = dtree.predict(x_test)

# Calculate the accuracy of the model
accuracy = dtree.score(x_test, y_test) * 100

print(f"Accuracy: {accuracy:.2f}%")
```

```python
# Load the Iris dataset
FlowerData = datasets.load_iris()

# Create a DataFrame with the dataset's features, ensuring column names are consistent
df = pd.DataFrame(FlowerData.data, columns=[name.lower() for name in FlowerData.feature_names])

# Get the target values
y = FlowerData.target

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=42)
```

```python
RF=RandomForestClassifier(n_estimators = 10, n_jobs = -1)

RF.fit(x_train, y_train)
```

```python
# testing the trained model
RF.predict(x_test)
  
# Calculate the accuracy of the model 
print("Accuracy:", RF.score(x_test, y_test)*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(RF,x_test,y_test,display_labels=FlowerData.target_names,cmap='Blues')
```

```python
import pandas as pd
from sklearn.datasets import load_iris
# Load the Iris dataset
FlowerData = datasets.load_iris()

# Create a DataFrame with the dataset's features, ensuring column names are consistent
df = pd.DataFrame(FlowerData.data, columns=[name.lower() for name in FlowerData.feature_names])
df = pd.DataFrame(FlowerData.data, columns=FlowerData.feature_names)
df[' Species'] = FlowerData.target
```

```python
# Check the column names
print(df.columns)

# Strip any leading/trailing spaces from the column names
df.columns = df.columns.str.strip()

# Verify the DataFrame structure
print(df.head())

# Split the dataset again
x_train, x_test, y_train, y_test = train_test_split(df[FlowerData.feature_names], df['Species'], random_state=42)
```

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Define and train the classifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

# Plot the decision tree
plt.figure(figsize=(20,10))  # Define the size of the plot (width, height)
tree.plot_tree(clf, filled=True, feature_names=FlowerData.feature_names, class_names=FlowerData.target_names)
plt.show()
```
